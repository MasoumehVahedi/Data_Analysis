# Machine Learning

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
path = 'D:/Projects for my GitLab/kaggle_project/weatherHistory.csv'
data = pd.read_csv(path)
display(data.head())
```

```python
# To assign numerical values to categorize the filed of Daily Summary class:
# First, I separate Daily Summary column in a pandas data frame
dailySummarry_df = pd.DataFrame(data['Daily Summary'], columns=['Daily Summary'])
dailySummarry_df
```

```python
# Second, creat an object and transform the variable into it assigned number to categorize
labelencoder = LabelEncoder()
dailySummarry_df['dailySummarry_types_cat'] = labelencoder.fit_transform(dailySummarry_df['Daily Summary'])
display(dailySummarry_df)
```

```python
# To add a new column for numerical variable
data = data.join(dailySummarry_df['dailySummarry_types_cat'])
display(data)
```

```python
# remove object columns
data.drop(['Daily Summary','Summary', 'Precip Type', 'Formatted Date', 'Loud Cover'], axis=1, inplace=True)
display(data)
```

```python
# Manging missing values
data.fillna(method='ffill', inplace=True)
```

```python
# Here we separate x and y which Daily Summary is our target to be predicted by weather parameters
x = data.drop(['dailySummarry_types_cat'], axis=1)
y = data['dailySummarry_types_cat']

# Separate test and training data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# Classification using Decision Tree model

# with gini

```python
# Set up the model using Gini Coefficient
model = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=15).fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("MSE: ", mean_squared_error(y_test, y_pred))
print("R2: ", model.score(x_test,y_test))
```

#  With Entropy

```python
# To fit the mpdel using Entropy index to compare with Gini coefficient
model2 = DecisionTreeClassifier(criterion="entropy", max_depth=None).fit(x_train,y_train)
y_pred = model2.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
